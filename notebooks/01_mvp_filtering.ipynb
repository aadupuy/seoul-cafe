```python
import pandas as pd
from src.recommender import recommend
```

```python
# Load the dataset
df = pd.read_csv('data/raw/cafes_seed.csv')
df.head()
```

```python
# Preprocess the text data for semantic search
df["tags"] = df["tags"].fillna("").str.lower()
df["category"] = df["category"].fillna("").str.lower()
df["district"] = df["district"].fillna("")

df["search_text"] = (
    df["name"] + " " +
    df["district"] + " " +
    df["category"] + " " +
    df["tags"]
).str.lower()
```

```python
# Run the recommendation function with a sample query
recommend(df, "quiet coffee")
```
